```python
import pandas as pd
import matplotlib.pyplot as plt

%store -r dt
```

```python
treineiros= dt.query('IN_TREINEIRO == 1')['NU_IDADE'].dropna()
n_treineiros= dt.query('IN_TREINEIRO == 0')['NU_IDADE'].dropna()

fig, (ax1, ax2)= plt.subplots(2, 1)

ax1.set_title('Treineiros')
ax1.set_xlabel('Idade')
ax1.set_ylabel('Participantes')

ax2.set_title('Não Treineiros')
ax2.set_xlabel('Idade')
ax2.set_ylabel('Participantes')

treineiros.hist(bins=100, ax=ax1)
n_treineiros.hist(bins=100, ax=ax2)

plt.tight_layout()
```

```python
treineiros= dt.query('IN_TREINEIRO == 1')['NU_IDADE'].dropna()
n_treineiros= dt.query('IN_TREINEIRO == 0')['NU_IDADE'].dropna()

plt.figure(figsize=(10,8))
plt.hist([treineiros, n_treineiros], bins = 100, label=['Treineiros', 'Não Treineiros'])
plt.title("Treineiros x  Não Treineiros")
plt.legend(loc='upper right')
plt.grid()

plt.savefig('txnt.png')
```
